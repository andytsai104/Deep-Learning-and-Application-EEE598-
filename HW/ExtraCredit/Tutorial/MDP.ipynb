# Markov Decision Process

- Terminology for Markov Decision Process: <br>
environment - the world or the system where the problem is<br>
agent - the one who's making actions and decisions that interact with the environment<br>
states - a set of possible situations in the environment, which the agent can experience<br>
actions - a set of available choices for the agent to choose<br>
reward - is the feedback after the agent take a certain action in a state<br>
policy - is a strategy that map states to actions, which maximize the long-term reward by guiding the agent's decision<br>

```python
def markov_process_step(state, transition_probabilities):
    probabilities = transition_probabilities[:, state]
    next_state = np.random.choice(len(probabilities), p=probabilities)
    return next_state
```

```python
def markov_reward_process_step(state, transition_probabilities, reward_struture):
    new_state = markov_process_step(state, transition_probabilities)
    reward = reward_struture[new_state]
    return new_state, reward
```

```python
def calculate_return(rewards, gamma):
    return_val = 0.0
    for t, reward in enumerate(rewards):
        return_val += reward * pow(gamma, t)
    return return_val
```

```python
def markov_decision_process_step_deterministic(state, transition_probabilities_given_action, reward_structure, policy):
    action = policy[state]
    probabilities = transition_probabilities_given_action[:, state, action]
    new_state = np.random.choice(len(probabilities), p=probabilities)
    reward = reward_structure[new_state]
    return new_state, reward
```
